```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import cmath
from matplotlib import cm
import random as rnd
import csv
import ast
import scipy as sp
import scipy.stats
```

```python
def brown_approx(n):
    nsteps= n
    dt= 1/float(nsteps)
    jumps= np.random.normal(0, 1, nsteps)
    Xs= np.zeros(nsteps)
    for i in range(nsteps-1): Xs[i+1]= Xs[i]+np.sqrt(dt)*jumps[i]
    x=np.linspace(0.0, 1.0, num=nsteps)
    return (x,Xs)
```

```python
xx, XX = brown_approx(10)
plt.plot(xx, XX)
xx, XX = brown_approx(20)
plt.plot(xx, XX)
xx, XX = brown_approx(100)
plt.plot(xx, XX)
xx, XX = brown_approx(1000)
plt.plot(xx, XX)
plt.show()
```
